## Linear Regression
Alasan memilih penggunaan Linear Regression adalah karena dapat menangkap hubungan linier antara fitur-fitur (seperti luas bangunan, jumlah kamar, lokasi, dll.) dan variabel target (harga perumahan) secara praktis. Kemudian, dapat diinterpretasikan dan memberikan pemahaman tentang bagaimana setiap fitur berkontribusi terhadap prediksi harga dan saling memengaruhi. Linear Regression bekerja dengan baik ketika hubungan antara fitur dan variabel target relatif sederhana dan linier. Selain itu, metode ini cenderung memiliki waktu training yang cepat.

```python
X = sales_numeric_no_outliers[["LotFrontage"]]
y = sales_numeric_no_outliers["LotArea"]
```

```python
# bagi data menjadi data train dan test

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 42)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
# gabungkan, ada berapa data test yang dimiliki
X_test.shape[0]/(X_train.shape[0] + X_test.shape[0])
```

```python
from sklearn.linear_model import LinearRegression
```

```python
# buat model machine learning linear regression
sales_lr = LinearRegression()
```

```python
# buat model training
sales_lr.fit(X_train, y_train)
```

```python
# dapatkan intercept
b0 = sales_lr.intercept_
print(b0)
```

```python
# dapatkan koefisien
b1 = sales_lr.coef_
print(b1)
```

```python
# masukkan ke dalam persamaan
# y = [koefisien] * X + intercept
y = [123.67858788] * X + 1128.0184539117836
```

```python
# testing manual
X_test.iloc[0].values
```

```python
y1 = b0 + X_test.iloc[0].values * b1
```

```python
y1
```

```python
sales_lr.predict(X_test)
```

```python
# rumus manual
y_pred_manual = b0 + b1 * X_test
```

```python
y_pred_manual
```

```python
sales_lr.predict(X_test)
```

## Evaluation

```python
y_pred = sales_lr.predict(X_test)
```

```python
# hitung error
error = (y_test - y_pred)
```

```python
error_sales = pd.DataFrame({"ground_truth": y_test,
              "prediction": y_pred,
              "error": error})
error_sales['error_pct'] = error_sales["error"] / error_sales["ground_truth"]
error_sales["error_sq"] = error_sales["error"] ** 2
```

```python
error.hist(bins = 100)
```

```python
error_sales.mean()
```

```python
mean_absolute_percentage_error = error_sales["error_pct"].abs().mean()
print(mean_absolute_percentage_error)
```

Dapat dilihat bahwa error yang didapatan hampir 26%. Ini menunjukkan bahwa masih terdapat 26% data yang masih belum bisa dilihat dengan benar.

```python
from sklearn.metrics import mean_squared_error as mse
from sklearn.metrics import mean_absolute_error as mae
from sklearn.metrics import mean_absolute_percentage_error as mape
```

```python
print("MSE Linear Regression: ",mse(y_test, y_pred))
print("MAE Linear Regression: ",mae(y_test, y_pred))
print("MAPE Linear Regression: ",mape(y_test, y_pred))
```

Pada linear regression mendapatkan nilai MSE menunjukkan bahwa rata-rata kuadrat dari selisih antara nilai prediksi dan nilai sebenarnya adalah sekitar 12333354.870864592. Nilai MAE menunjukkan bahwa rata-rata dari selisih absolut antara nilai prediksi dan nilai sebenarnya adalah sekitar 2277.4388921479726. Dapat dilihat nilai MAPE yang lebih rendah menunjukkan bahwa model memiliki kesalahan prediksi relatif yang lebih kecil. Dalam kasus ini, nilai MAPE menunjukkan bahwa rata-rata dari persentase selisih absolut antara nilai prediksi dan nilai sebenarnya adalah sekitar 26.26%.

## Decision Tree Regressor

Alasan memilih penggunaan Decision Tree adalah mudah menangani fitur-fitur numerik maupun kategorikal, yang sering ditemukan dalam dataset pertimbangan pemilihan rumah. Metode ini dapat menangkap pola-pola kompleks dalam data dan membagi ruang fitur menjadi wilayah-wilayah yang lebih kecil dan homogen. Hasil model metode ini mudah diinterpretasikan dan divisualisasikan. Selain itu, metode ini dapat menangani data yang memiliki banyak fitur dan outlier dengan baik.

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
sales_dt = DecisionTreeRegressor()
```

```python
sales_dt.fit(X_train, y_train)
```

```python
y_pred_sales_dt = sales_dt.predict(X_test)
```

```python
print("MSE Decision Tree: ",mse(y_test, y_pred_sales_dt))
print("MAE Decision Tree: ",mae(y_test, y_pred_sales_dt))
print("MAPE Decision Tree: ",mape(y_test, y_pred_sales_dt))
```

Pada decision tree mendapatkan nilai MSE menunjukkan bahwa rata-rata kuadrat dari selisih antara nilai prediksi dan nilai sebenarnya adalah sekitar 14210775.461613953. Nilai MAE menunjukkan bahwa rata-rata dari selisih absolut antara nilai prediksi dan nilai sebenarnya adalah sekitar 2269.6194369670147. Dapat dilihat nilai MAPE yang lebih rendah menunjukkan bahwa model memiliki kesalahan prediksi relatif yang lebih kecil. Dalam kasus ini, nilai MAPE menunjukkan bahwa rata-rata dari persentase selisih absolut antara nilai prediksi dan nilai sebenarnya adalah sekitar 23.54%.

Setelah dilakukan perbandingan antara penggunaan Linear Regression dan Decision Tree, dapat diketahui bahwa model pada metode Decision Tree memiliki hasil yang lebih baik.
